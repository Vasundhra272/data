```python
Species=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colour=['Red','Blue','Orange']


for  i in range(3):
    x = iris[iris['Species'] == Species[i]]['SepalLengthCm']
    plt.scatter(x, [i] * len(x), color=colour[i], label=Species[i])

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Species (Represented by index)")
plt.title("Sepal Length Distribution for Iris Species")
plt.legend()
plt.show()
```

```python
Species=['Iris-setosa','Iris-versicolor','Iris-virginica']
color=['Red','Blue','orange']
for i in range(3):
  x=iris[iris['Species']==Species[i]]
  plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'],c=color[i],label=Species[i])
  plt.legend()
```

```python
iris['Species']=iris['Species'].astype('category')
iris['Species']=iris['Species'].cat.codes
```

```python
iris['Species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x=iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=iris['Species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

Accuracy



```python
model.score(x,y)*100
```
